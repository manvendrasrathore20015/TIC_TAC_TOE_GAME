```python
player=None
board=["#","#","#","#","#","#","#","#","#"]

def display_board():
    print("\n  {} | {} | {}  ".format(board[6],board[7],board[8]))
    print("-------------")
    print("  {} | {} | {}  ".format(board[3],board[4],board[5]))
    print("-------------")
    print("  {} | {} | {}  \n".format(board[0],board[1],board[2]))

def pl_choice():
    global player
    input_no=True
    print("Player {} turn".format(player),end=" ")
    
    while input_no:
        choice=int(input("enter your position from 1-9 -> "))
        if(board[choice-1]=="#"):
            board[choice-1]=player
            display_board()
            input_no=False
        else:
            print("position already choosen")
        
def flip_player():
    global player
    if (player=="X"):
        player="O"
    elif (player=="O"):
        player="X"
    
def check_won():
    global applicable

    won=False
    if (board[0]==board[1]==board[2]=="X" or board[3]==board[4]==board[5]=="X" or board[6]==board[7]==board[8]=="X"):
        print("{} won".format(player))
        won=True
    elif (board[0]==board[3]==board[6]=="X" or board[1]==board[4]==board[7]=="X" or board[2]==board[5]==board[8]=="X"):
        print("{} won".format(player))
        won=True
    elif (board[0]==board[4]==board[8]=="X" or board[6]==board[4]==board[2]=="X"):
        print("{} won".format(player))
        won=True
    elif (board[0]==board[1]==board[2]=="O" or board[3]==board[4]==board[5]=="O" or board[6]==board[7]==board[8]=="O"):
        print("{} won".format(player))
        won=True
    elif (board[0]==board[3]==board[6]=="O" or board[1]==board[4]==board[7]=="O" or board[2]==board[5]==board[8]=="O"):
        print("{} won".format(player))
        won=True
    elif (board[0]==board[4]==board[8]=="O" or board[6]==board[4]==board[2]=="O"):
        print("{} won".format(player))
        won=True
        
    if(won==True):
        applicable=False
        again=input("PRESS 'Y' to play again and 'N' to quit")
        if(again=="Y"):
            play_game()
        elif(again=="N"):
            print("The game is over")
    
def check_tie():
    tie=False
    global applicable
    count=0
    
    for i in range(0,9):
        if(board[i]=="#"):
            count+=1
    if(count==0):
        print("It is a tie")
        applicable=False
        tie=True
        
    if(tie==True):
        again=input("PRESS 'Y' to play again and 'N' to quit")
        if(again=="Y"):
            play_game()
        elif(again=="N"):
            print("The game is over")
  
            
def play_game():
    global applicable
    global board
    global player
    applicable=True
    playing=True
    board=["#","#","#","#","#","#","#","#","#"]
    display_board() 
    
    while playing:
        player=input("choose between 'X' and 'O' :--> ")
        if (player=="X" or player=="O"):
            playing=False
            break;
        else:
            print("enter correct choice")

    while applicable:
        pl_choice()
        check_won()
        check_tie()
        flip_player()
            
play_game()
```
